```python
import h5py
import numpy as np
import scipy.io

def leer_v73(archivo, var):
    with h5py.File(archivo, 'r') as f:
        data = f[var]
        
        # Intentamos acceder a los campos 'real' e 'imag' directamente.
        # Si no existen, esto lanzará un KeyError y pasamos al bloque 'except'
        # para leer el dato como una matriz normal.
        try:
            real_part = np.array(data['real']).T
            imag_part = np.array(data['imag']).T
            return real_part + 1j * imag_part
        except (KeyError, ValueError):
            # Si 'real'/'imag' no existen, leemos el dataset directamente
            return np.array(data).T

def procesar_matrices_v73(archivo_real, archivo_imag, nombre_var):
    print("Leyendo archivos...")
    
    # 1. Leer las partes
    parte_r = leer_v73(archivo_real, nombre_var)
    parte_i = leer_v73(archivo_imag, nombre_var)

    # 2. Combinar en complejo
    matriz_compleja = parte_r.astype(np.complex128) + 1j * parte_i.astype(np.complex128)
    
    print(f"Matriz combinada. Dimensiones: {matriz_compleja.shape}")

    # 3. Guardar
    np.save('matriz_final.npy', matriz_compleja)
    scipy.io.savemat('matriz_final.mat', {'matriz_resultado': matriz_compleja})
    print("Archivos guardados con éxito.")

# --- CONFIGURACIÓN ---
archivo_r = 'campo02_test.mat'
archivo_i = 'campo01_test.mat'
nombre_variable = 'beam'

if __name__ == "__main__":
    procesar_matrices_v73(archivo_r, archivo_i, nombre_variable)
```

```python
import h5py

def inspeccionar_archivo(archivo):
    with h5py.File(archivo, 'r') as f:
        print(f"--- Contenido de {archivo} ---")
        # Listamos las llaves principales
        for key in f.keys():
            print(f"Llave encontrada: {key}")
            # Si hay sub-objetos, los mostramos
            try:
                print(f"  Contenido de {key}: {list(f[key])}")
            except:
                pass

inspeccionar_archivo('planewave_256.mat')
```
